```python
characters = [
    'MERCUTIO',
    'PARIS',
    'MONTAGUE',
    'LADY MONTAGUE',
    'ROMEO',
    'BENVOLIO',
    'ABRAM',
    'BALTHASAR',
    'CAPULET',
    'LADY CAPULET',
    'JULIET',
    'TYBALT',
    'CAPULET’S COUSIN',
    'NURSE',
    'PETER',
    'SAMPSON',
    'GREGORY',
    'FRIAR LAWRENCE',
    'FRIAR JOHN'
]
    
```

```python
len(acts)
```

```python
charnum = {}

connections = {}
for i in range(len(characters)-1):
    for j in range(i+1,len(characters)):
        connections[(characters[i],characters[j])] = 0

for k in characters:
    charnum[k] = 0

for i in acts:
    for j in i.split('SCENE')[1:]:
        scenechars = []
        for k in characters:
            if j.find(k) != -1:
                scenechars.append(k)
                charnum[k] += 1
        for a in range(len(scenechars)-1):
            for b in range(a+1,len(scenechars)):
                connections[(scenechars[a],scenechars[b])] += 1
```

The following are all the characters in the play and the number of scenes that they appear in. This information will be used as notes in the later visualization.

The following code creates the nodes for the visualization.

```python
graphitems = []

for k,v in charnum.items():
    dashnode = {'data': {'id': k,
                         'label': k.title(),
                         'size': str(v)}}
    graphitems.append(dashnode)
    
graphitems
```

```python
graphitems = []

for k,v in charnum.items():
    dashnode = {'data': {'id': k,
                         'label': k.title(),
                         'size': str(v)}}
    graphitems.append(dashnode)

for k,v in connections.items():
    if v != 0:
        dashedge = {'data': {'source': k[0],
                             'target': k[1],
                             'weight': v}}
        graphitems.append(dashedge)
    
graphitems
```
