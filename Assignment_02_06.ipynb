```python
response = requests.get('https://corgis-edu.github.io/corgis/csv/')
```

```python
# use BeautifulSoup to conduct a search
from bs4 import BeautifulSoup
```

```python
# Save all the html in a string variable
html_string = response.text

# Use BeautifulSoup to create a new object that will allow you to search for HTML tags
document = BeautifulSoup(html_string, "html.parser")
```

```python
# In order to be able to make a list of all CORGIS datasets, use "find_all" to look for HTML tag with 
#headings (h3)
document.find_all('h3')
```

```python
corgisdatasets = document.find_all('h3')
```

```python
def pdcorgis(a):
    a = a.lower()
    response = requests.get('https://corgis-edu.github.io/corgis/csv/' + a + '/')
    html_string = response.text
    document = BeautifulSoup(html_string, "html.parser")
    document.find_all('a')
    document2 = document.find_all('a')
    for i in document2:
        if i.has_attr('download') == True:
            link = 'https://corgis-edu.github.io/corgis/csv/' + a + '/' + i['href']
            corgisdf = pd.read_csv(link)
            return corgisdf
            
```

```python
df = pdcorgis("Classics")
```

```python
df.head()
```

```python
df.sort_values(by='metrics.statistics.words',ascending=True)[-20:].plot.barh(x='bibliography.title',y='metrics.statistics.words')
```
